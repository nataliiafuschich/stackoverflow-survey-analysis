10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
df_for_ind = df[['ConvertedCompYearly', 'RemoteWork', 'Industry']]
df_for_ind
```

```python
df_for_ind = df_for_ind.dropna(subset=['ConvertedCompYearly'])
```

```python
perc_75 = df_for_ind['ConvertedCompYearly'].quantile(0.75)
perc_75
```

```python
high_paid_remote = df_for_ind[(df_for_ind['ConvertedCompYearly'] > perc_75) & (df_for_ind['RemoteWork'] == 'Remote')]
```

```python
high_paid_remote
```

```python
industry = high_paid_remote['Industry'].value_counts().reset_index()
industry
```

Висновок: Серед високооплачуваних респондентів, які працюють віддалено, найпоширенішими індустріями є Software Development, Fintech та Healthcare
